```python
# normaliser les variables quantitatives chaque valeur/valeur max
def normalize(dataframe):
    for i in dataframe.columns:
        if(i in dataframe.select_dtypes("object").columns)==False:
            dataframe[i]=dataframe[i]/dataframe[i].max()
    return dataframe
normalize(df)
```

```python
#recorder variable qualitative
def recoder(column): 
    return column.astype('category').cat.codes
```

```python

#permet de recoder toutes les variables de type objet
def recoder_df(data):
    for col in data.select_dtypes('object').columns:
        data[col] = recoder(data[col])
#serie.astype('category').cat.codes

    return data

recoder_df(df)
```

```python
#créer un dataframe x qui va contenir de age a pente et y coeur
x=df.iloc[:,:-1]
y=df.CŒUR
```

```python
#train_split() fonction on va lui donner en paramètre  x et y et le taux 0.3
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
#(regression logistique avec Linear_model)
model=LogisticRegression(max_iter=1000 ,random_state=42)
#Entrainement du model
model.fit(X_train,y_train)
#Retourne la performance
model.score(X_train,y_train)
```
